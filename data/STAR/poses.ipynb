```python
from star.ch.star import STAR
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import torch
label_list=['female','male','neutral']
label=label_list[2]
model = STAR(gender=label,num_betas=10)
## Assign random pose and shape parameters
#model.pose[:] = np.random.rand(model.pose.size) * 1.
model.pose[:] = np.zeros(model.pose.size) * 1.
# model.betas[:] = np.random.rand(model.betas.size) * .03
for j in range(0,10):
    model.betas[:] = 0.0  #Each loop all PC components are set to 0.
    fig = plt.figure(figsize=(2*800/72,800/72))    
    for n,i in enumerate(np.linspace(-4,4,5)): #Varying the jth component +/- 3 standard deviations
        ax = fig.add_subplot(1, 5, n+1, projection='3d')
        model.betas[j] = i        
        x=model.r[:,0]
        y=model.r[:,1]
        z=model.r[:,2]
        ax.scatter3D(x, z, y, c=y*np.sqrt(x**2+z**2), cmap='bone')
        ax.set_xlim([-1,1]);ax.set_ylim([-1,1]);ax.set_zlim([-1,1])
        ax.view_init(10, 45)
    plt.show()
```
